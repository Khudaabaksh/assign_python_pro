```python
def arm(num):
    n=num
    s=0
    while(n>0):
        p=n%10
        s=s+(p*p*p)
        n=n//10
    if num==s:
        print("{} is armstrong".format(num))
    else:
        print("{} is not armstrong".format(num))
arm(407)
```

```python
print("from where you want")
a=int(input())
print("till where")
b=int(input())
for i in range(a,b):
    arm(i)
```
